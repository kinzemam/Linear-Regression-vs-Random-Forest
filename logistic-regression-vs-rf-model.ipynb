```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
```

```python
df.drop(columns=['Unnamed: 32'], inplace=True)
```

```python
df.loc[df.diagnosis=='B', 'diagnosis_num']=0
df.loc[df.diagnosis=='M', 'diagnosis_num']=1
```

```python
x=df.drop(columns=['id', 'diagnosis', 'diagnosis_num'])
y=df['diagnosis_num']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(x_train_scaled, y_train)
```

```python
y_prob_lr=lr_model.predict_proba(x_test_scaled, )[:, 1]
y_pred_lr=(y_prob_lr>=0.5).astype(int)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
```

```python
y_prob_rf=rf_model.predict_proba(x_test)[:,1]
y_pred_rf=rf_model.predict(x_test)
```

# Model Comparision

```python
from sklearn.metrics import classification_report
print("Logistic Regression Report: ")
lr_report=classification_report(y_test, y_pred_lr)
print(lr_report)
rf_report=classification_report(y_test, y_pred_rf)
print("Random Forest Report: ")
print(rf_report)
```

# ROC and AUC

## 1. logitic regression

```python
from sklearn.metrics import roc_curve, roc_auc_score
fpr_lr, tpr_lr, thr_lr=roc_curve(y_test, y_prob_lr)
lr_roc_auc=roc_auc_score(y_test, y_prob_lr).round(4)
plt.figure(figsize=(8,6))
plt.plot(fpr_lr, tpr_lr, label=f'AUC= {lr_roc_auc}')
plt.xlabel("False Positive Rate (1-specificity)")
plt.ylabel('True Positive Rate (sensitivity)')
plt.title("ROC curve for Logistic Regression")
plt.legend()
```

## 2. random forest

```python
fpr_rf, tpr_rf, thr_rf=roc_curve(y_test, y_prob_rf)
rf_roc_auc=roc_auc_score(y_test, y_prob_rf).round(4)
plt.figure(figsize=(8,6))
plt.plot(fpr_rf, tpr_rf, label=f'AUC= {rf_roc_auc}')
plt.xlabel("False Positive Rate (1-specificity)")
plt.ylabel('True Positive Rate (sensitivity)')
plt.title("ROC curve for Random Forest")
plt.legend()
```

## ROC Curve and AUC Analysis

The ROC (Receiver Operating Characteristic) curve illustrates the trade-off between
True Positive Rate (Recall) and False Positive Rate at various classification thresholds.
The Area Under the Curve (AUC) summarizes the model’s ability to distinguish between
classes.

### Logistic Regression
- AUC score ≈ **0.997**
- Indicates excellent class separation and strong probabilistic predictions.

### Random Forest
- AUC score ≈ **0.995**
- Also demonstrates very strong performance, though slightly lower than Logistic Regression.

### Result
Both models perform exceptionally well in terms of ROC–AUC. Logistic Regression shows
a marginally higher AUC, suggesting better overall ranking of positive vs negative cases.
However, Random Forest remains highly competitive and robust.


# Conclusion

Both Logistic Regression and Random Forest models were evaluated using the default classification threshold of 0.5 to ensure a fair and consistent comparison.

Logistic Regression achieved slightly better overall performance with an accuracy of 97% compared to 96% for Random Forest. For the heart disease class, Logistic Regression demonstrated higher recall (0.95 vs 0.93) while maintaining equally strong precision.

Although Random Forest effectively captured non-linear patterns, Logistic Regression provided comparable or better performance while remaining simpler and more interpretable. This makes Logistic Regression a strong baseline model for this dataset, particularly suitable for medical applications where interpretability and recall are important.
