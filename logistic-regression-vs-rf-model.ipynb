```python
df.drop(columns=['Unnamed: 32'], inplace=True)
```

```python
df.loc[df.diagnosis=='B', 'diagnosis_num']=0
df.loc[df.diagnosis=='M', 'diagnosis_num']=1
```

```python
x=df.drop(columns=['id', 'diagnosis', 'diagnosis_num'])
y=df['diagnosis_num']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(x_train_scaled, y_train)
```

```python
y_prob_lr=lr_model.predict_proba(x_test_scaled, )[:, 1]
y_pred_lr=(y_prob_lr>=0.5).astype(int)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
```

```python
y_pred_rf=rf_model.predict(x_test)
```

# Model Comparision

```python
from sklearn.metrics import classification_report
print("Logistic Regression Report: ")
lr_report=classification_report(y_test, y_pred_lr)
print(lr_report)
rf_report=classification_report(y_test, y_pred_rf)
print("Random Forest Report: ")
print(rf_report)
```

# Conclusion

Both Logistic Regression and Random Forest models were evaluated using the default classification threshold of 0.5 to ensure a fair and consistent comparison.

Logistic Regression achieved slightly better overall performance with an accuracy of 97% compared to 96% for Random Forest. For the heart disease class, Logistic Regression demonstrated higher recall (0.95 vs 0.93) while maintaining equally strong precision.

Although Random Forest effectively captured non-linear patterns, Logistic Regression provided comparable or better performance while remaining simpler and more interpretable. This makes Logistic Regression a strong baseline model for this dataset, particularly suitable for medical applications where interpretability and recall are important.
